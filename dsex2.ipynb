```python
import pandas as pd
```

```python
df=pd.read_csv('housing_dataset.csv')
```

```python
#Handling Null Values
#Identify the columns with null values in the dataset.
```

```python
df.isna().sum()
```

```python
#Impute the missing values in the &#39;Price&#39; column with the median price.
```

```python
mid=df['Price'].median()
mid
```

```python
df['Price'] = df['Price'].fillna(mid)
df
```

```python
#Impute the missing values in the &#39;Bathrooms&#39; column with the mean number of
#bathrooms.
```

```python
mn=df['Bathrooms'].mean()
mn
```

```python
df['Bathrooms']=df['Bathrooms'].fillna(mn)
df
```

```python
#Drop any rows that still contain null values after imputation.
```

```python
df=df.dropna(axis=0)
df
```

```python
#Handling Outliers
#Identify outliers in the &#39;Price&#39; column using the Interquartile Range (IQR) method.
```

```python
q1=df['Price'].quantile(0.25)
q2=df['Price'].quantile(0.75)
iqr=q2-q1
lb=q1-1.5*iqr
ub=q2+1.5*iqr
outlier=(df['Price']<lb) | (df['Price']>ub)
#Remove the outliers from the dataset
cleaned=df[~outlier]
cleaned
```

```python
#Normalize the &#39;Sqft_Living&#39; column using Min-Max scaling after removing outliers.
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
df['Sqft_Living_Normalized']=scaler.fit_transform(df[['Sqft_Living']])
df
```
